## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
lendingDataDF = pd.read_csv("lending_data.csv")

# Review the DataFrame
lendingDataDF.head()
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y=lendingDataDF["loan_status"]

# Separate the X variable, the features
X=lendingDataDF.drop(columns="loan_status")
```

```python
# Review the y variable Series
y.head()
```

```python
# Review the X variable DataFrame
X.head()
```

### Step 3: Check the balance of the labels variable (`y`) by using the `value_counts` function.

```python
# Check the balance of our target values
y.value_counts()
```

### Step 4: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
logresmodel = LogisticRegression(random_state=1)

# Fit the model using training data
logresmodel.fit(X_train, y_train)
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
y_pred =logresmodel.predict(X_test)
```

### Step 3: Evaluate the model’s performance by doing the following:

* Calculate the accuracy score of the model.

* Generate a confusion matrix.

* Print the classification report.

```python
# Print the balanced_accuracy score of the model
from sklearn.metrics import balanced_accuracy_score
print(f"Logistic regression model balanced accuracy: {balanced_accuracy_score(y_test, y_pred):.3f}")
```

```python
# Generate a confusion matrix for the model
confusion_matrix(y_test, y_pred)
```

```python
# Print the classification report for the model
print(classification_report(y_test, y_pred))
```

### Step 4: Answer the following question.

**Question:** How well does the logistic regression model predict both the `0` (healthy loan) and `1` (high-risk loan) labels?

**Answer:** The logistic regression model predicted the healthy loans almost perfectly, with 100% precision and 99% recall.

 While the logistic regression model wasn't quite as accurate at predicting the high-risk loans (likely due to the data containing approximately 30x as many healthy loans than high-risk loans), it was also very good at predicting the high-risk loans, with 84% precision and 94% recall.

 This indicates that the logistic regression model is an excellent predictor of healthy loans, and a very good predictor of high-risk loans.

## Predict a Logistic Regression Model with Resampled Training Data

### Step 1: Use the `RandomOverSampler` module from the imbalanced-learn library to resample the data. Be sure to confirm that the labels have an equal number of data points.

```python
# Import the RandomOverSampler module form imbalanced-learn (using this as a sub for test-train-score)
from imblearn.over_sampling import RandomOverSampler

# Instantiate the random oversampler model

# # Assign a random_state parameter of 1 to the model use
rosmodel = RandomOverSampler(random_state=1)

# Fit the original training data to the random_oversampler model/ use .fit
X_resampled, y_resampled = rosmodel.fit_resample(X_train, y_train)
```

```python
# Count the distinct values of the resampled labels data
y_resampled.value_counts()
```

### Step 2: Use the `LogisticRegression` classifier and the resampled data to fit the model and make predictions.

```python
# Instantiate the Logistic Regression model
from sklearn.linear_model import LogisticRegression

# Assign a random_state parameter of 1 to the model
logresmodel = LogisticRegression(random_state=1)

# Fit the model using the resampled training data
logresmodel.fit(X_resampled, y_resampled)

# Make a prediction using the testing data
y_repred =logresmodel.predict(X_test)
```

### Step 3: Evaluate the model’s performance by doing the following:

* Calculate the accuracy score of the model.

* Generate a confusion matrix.

* Print the classification report.

```python
# Print the balanced_accuracy score of the model
from sklearn.metrics import balanced_accuracy_score
print(f"Logistic regression model balanced accuracy: {balanced_accuracy_score(y_test, y_repred):.3f}")
```

```python
# Generate a confusion matrix for the model
confusion_matrix(y_test, y_repred)
```

```python
# Print the classification report for the model
print(classification_report(y_test, y_repred))
```

### Step 4: Answer the following question

**Question:** How well does the logistic regression model, fit with oversampled data, predict both the `0` (healthy loan) and `1` (high-risk loan) labels?

**Answer:** As with the model fitted with the original data, the logistic regression model fit with oversampled data predicted the healthy loans almost perfectly, with 100% precision and 99% recall.

 While the logistic regression model wasn't quite as accurate at predicting the high-risk loans (likely due to the data containing approximately 30x as many healthy loans than high-risk loans), it was also very good at predicting the high-risk loans, with 84% precision and 99% recall. **This represents a 5% increase in recall from the high-risk loans in the original model.**

 This indicates that the logistic regression model fit with oversampled data is an excellent predictor of healthy loans, and a very good predictor of high-risk loans.
